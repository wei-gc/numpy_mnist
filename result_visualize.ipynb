```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_history = np.load('results/train_history.npy', allow_pickle=True).item()
```

```python
train_loss = train_history['train_loss']
test_loss = train_history['test_loss']
test_acc = train_history['test_acc']
```

```python
plt.plot(train_loss, label='train_loss')
plt.plot(test_loss, label='test_loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()

plt.savefig('results/loss_curve.png', dpi=300, bbox_inches='tight', pil_kwargs={'compression': 'tiff_lzw'})
```

```python
plt.plot(np.array(test_acc)*100)
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.savefig('results/acc_curve.png', dpi=300, bbox_inches='tight', pil_kwargs={'compression': 'tiff_lzw'})
```

```python
network_params = np.load('results/params.npy', allow_pickle=True).item()
W1 = network_params['W1']
W2 = network_params['W2']
b1 = network_params['b1']
b2 = network_params['b2']
```

```python
plt.figure(figsize=(8, 16))
plt.subplot(2, 1, 1)
plt.pcolor(W1)
plt.colorbar()
plt.title('W1')
plt.subplot(2, 1, 2)
plt.pcolor(W2)
plt.colorbar()
plt.title('W2')
plt.savefig('results/weight.png', dpi=300, bbox_inches='tight', pil_kwargs={'compression': 'tiff_lzw'})
```

```python
plt.figure(figsize=(2, 8))
plt.subplot(2, 1, 1)
plt.pcolor(b1[:,None])
plt.colorbar()
plt.title('b1')

plt.subplot(2, 1, 2)
plt.pcolor(b2[:,None])
plt.colorbar()
plt.title('b2')
plt.savefig('results/bias.png', dpi=300, bbox_inches='tight', pil_kwargs={'compression': 'tiff_lzw'})
```
